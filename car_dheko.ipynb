```python
import json

# Define helper functions to extract car details
def car_overview_dict_creator(overview):
    overview_dict = json.loads(overview.replace("'", '"').replace('None', "null"))
    return {item['key']: item['value'] for item in overview_dict.get('top', [])}

def car_feature_dict_creator(feature):
    all_features = {}
    feature_dict = json.loads(feature.replace("'", '"').replace('None', "null"))
    # Extract 'top' features
    for dictionary in feature_dict.get('top', []):
        all_features[dictionary['value']] = True
    # Extract nested 'data' features
    for section in feature_dict.get('data', []):
        for item in section.get('list', []):
            all_features[item['value']] = True
    return all_features

def car_spec_dict_creator(spec):
    specs_dict = json.loads(spec.replace("'", '"').replace('None', "null"))
    all_specs = {item['key']: item['value'] for item in specs_dict.get('top', [])}
    # Extract nested specs from 'data'
    for dictionary in specs_dict.get('data', []):
        for item in dictionary.get('list', []):
            all_specs[item['key']] = item['value']
    return all_specs

# Load the Excel file
df = pd.read_excel('car_details_by_citys.xlsx')

# Extract car details
df['new_car_detail'] = df['new_car_detail'].apply(lambda x: json.loads(x.replace("'", '"').replace('None', "null")))
df_car_detail = pd.DataFrame(df['new_car_detail'].tolist())

# Extract car overview
df['new_car_overview'] = df['new_car_overview'].apply(car_overview_dict_creator)
df_car_overview = pd.DataFrame(df['new_car_overview'].tolist())

# Extract car features
df['feature_dict'] = df['new_car_feature'].apply(car_feature_dict_creator)
df_car_feature = pd.json_normalize(df['feature_dict']).fillna(False).astype(int)

# Extract car specs
df['new_car_specs'] = df['new_car_specs'].apply(car_spec_dict_creator)
df_car_specs = pd.DataFrame(df['new_car_specs'].tolist())

# Combine DataFrames and save to Excel
df_main = df[['city', 'car_links']]
final_df = pd.concat([df_main, df_car_detail, df_car_overview, df_car_feature, df_car_specs], axis=1)
final_df.to_excel('extracted_car_details.xlsx', index=False)

print("Data extracted and saved successfully as 'extracted_car_details.xlsx'")
```

```python
#Convert price to float:
def price_convert(price_str):
    price_str = price_str.replace('₹', '').replace(',', '').replace('â‚¹', '').strip()
    if 'Lakh' in price_str:
        return float(price_str.replace('Lakh', '').strip()) * 100000
    elif 'Thousand' in price_str:
        return float(price_str.replace('Thousand', '').strip()) * 1000
    elif 'Crore' in price_str:
        return float(price_str.replace('Crore', '').strip()) * 10000000
    else:
        # Return as float if no unit is found
        return float(price_str)

# Apply the function to the Price column
Final['price'] = Final['price'].apply(price_convert)

# Display the first few rows to confirm changes >900000
Final[Final['price']>900000]
```

```python
import re
import pandas as pd

# Define regex patterns for each column's units
unit_patterns = {
    'Engine_Displacement': r'CC',
    'Mileage': r'kmpl',
    'Length': r'mm',
    'Width': r'mm',
    'Height': r'mm',
    'Wheel Base': r'mm',
    'Front Tread': r'mm',
    'Rear Tread': r'mm',
    # 'BoreX Stroke': r'mm',
    'Ground Clearance Unladen': r'mm',
    'Kerb Weight': r'kg|kgs',
    'Gross Weight': r'kg|kgs',
    'Turning Radius': r'metres',
    'Cargo Volumn': r'-litres|Litre|Lit',
    'Acceleration': r'Seconds|seconds|Sec',
    'Top Speed': r'kmph|Kmph'
}

# Function to clean, convert columns and handle NaN values by replacing with median
def clean_and_convert_column_units(df):
    for col, pattern in unit_patterns.items():
        if col in df.columns:
            if df[col].dtype == 'object':  # Ensure column is of type object (string)
                # Remove units
                df[col] = df[col].astype(str).str.replace(pattern, '', regex=True)
                df[col] = pd.to_numeric(df[col], errors='coerce')  # Convert to numeric (float), coerce errors to NaN

            # Handle NaN values by replacing with the median of the column
            if df[col].isna().any():  # Check for NaN values
                median_value = df[col].median()  # Calculate the median of the column
                df[col] = df[col].fillna(median_value)  # Replace NaN with the median value

            # Convert to int if the column is listed for integer conversion
            if col in ['Engine_Displacement', 'Length', 'Width', 'Height', 'Wheel Base', 'Front Tread', 'Rear Tread', 'Kerb Weight', 'Gross Weight', 'Top Speed']:
                df[col] = df[col].astype(int)  # Convert to int after handling NaN
            
    return df

# Apply the function to clean, convert columns, and replace NaN with median
Final = clean_and_convert_column_units(Final)

Final
```

# **3 Handling Missing Values and Encoding**

```python
# Separate numeric and object columns
numeric_columns = Final.select_dtypes(include=['number']).columns
object_columns = Final.select_dtypes(include=['object']).columns

# Fill null values in numeric columns with the mean
for column in numeric_columns:
    Final[column].fillna(Final[column].mean(), inplace=True)

# Fill null values in object columns with the mode (most frequent value)
for column in object_columns:
    Final[column].fillna(Final[column].mode()[0], inplace=True)

# Save the cleaned DataFrame to a CSV file
Final.to_csv('car_dheko_filled.csv', index=False)

Final
```

RFE iteratively removes the least important features based on a model’s performance.

```python
# The error will come because X contains NaN values, and LinearRegression does not support missing values by default. 
# To resolve this, you can handle missing values by either filling them in (imputation) or dropping rows with NaNs.

# Here's how to handle the NaN values before applying RFE:
```

```python
# Impute Missing Values

# use an imputer to fill in the missing values. For instance, numerical features can be filled with the mean, median, or other statistics.

from sklearn.impute import SimpleImputer
from sklearn.pipeline import make_pipeline
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression  # Replace with your desired estimator

# Define the estimator for RFE (e.g., Linear Regression)
lr = LinearRegression()

# Create a pipeline to handle missing values and perform RFE
imputer = SimpleImputer(strategy='mean')  # You could use 'median' or 'most_frequent' as well
pipeline = make_pipeline(imputer, RFE(estimator=lr, n_features_to_select=10))

# Apply the pipeline
pipeline.fit(X, y)

# Extract selected features
rfe_step = pipeline.named_steps['rfe']
selected_features = X.columns[rfe_step.support_]
print("Selected features using RFE with imputation:")
print(selected_features)
```

# **5 Removing Outliers and Model Development**

```python
# Remove outliers using IQR method, including 'price'
def remove_outliers_iqr(car_dheko_Final, columns):
    for col in columns:
        if col in car_dheko_Final.columns:
            Q1 = car_dheko_Final[col].quantile(0.10)
            Q3 = car_dheko_Final[col].quantile(0.90)
            IQR = Q3 - Q1
            lower_bound = Q1 - 1.5 * IQR
            upper_bound = Q3 + 1.5 * IQR
            car_dheko_Final = car_dheko_Final[(car_dheko_Final[col] >= lower_bound) & (car_dheko_Final[col] <= upper_bound)]
    return car_dheko_Final

# Specify the columns for outlier removal, including 'price'
outlier_columns = numerical_features + ['price']  # Assuming 'numerical_features' is defined
cd = remove_outliers_iqr(car_dheko_Final, outlier_columns)

# Output the cleaned data
cd
```

Model Development

```python
# Step 1: Import Necessary Libraries
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

# Step 2: Define Features and Target Variable
# Define features and target
features = ['city', 'oem', 'model', 'modelYear', 'Kms_Driven', 'Fuel Type', 'Ownership', 'Transmission', 'Max Power',
            'Engine Type', 'Mileage', 'Seating_Capacity', 'Engine_Displacement', 'body_type', 'Acceleration']  #  variantName, Insurance Validity, Length 

X = cd[features]  # Select the features from the DataFrame
y = cd['price']   # Define the target variable

# Step 3: Split the Dataset into Training and Testing Sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)  # 70-30 split

# Step 4: Initialize Models
models = {
    'Linear Regression': LinearRegression(),
    'Decision Tree': DecisionTreeRegressor(random_state=42),
    'Random Forest': RandomForestRegressor(random_state=42),
    'Gradient Boosting': GradientBoostingRegressor(random_state=42)
}

# Step 5: Train and Evaluate Models Using Cross-Validation
for model_name, model in models.items():
    cv_scores = cross_val_score(model, X_train, y_train, cv=5, scoring='neg_mean_squared_error')
    print(f"{model_name} - Mean CV MSE: {-np.mean(cv_scores)}")

# Step 6: Fit and Predict with Each Model
for model_name, model in models.items():
    model.fit(X_train, y_train)  # Train the model
    y_pred = model.predict(X_test)  # Predict on test set
    
    # Evaluate the model
    print(f"\n{model_name} Performance:")
    print(f"Mean Absolute Error: {mean_absolute_error(y_test, y_pred)}")
    print(f"Mean Squared Error: {mean_squared_error(y_test, y_pred)}")
    print(f"R2 Score: {r2_score(y_test, y_pred)}")
```

```python
# Step 7: Hyperparameter Tuning with Grid Search (Example for Random Forest)
param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10]
}

grid_search = GridSearchCV(RandomForestRegressor(random_state=42), param_grid, cv=5, scoring='neg_mean_squared_error')
grid_search.fit(X_train, y_train)

# Best Parameters and Best Score
print("\nBest Parameters (Random Forest):", grid_search.best_params_)
print("Best CV Score:", -grid_search.best_score_)

# Step 8: Evaluate the Tuned Model on Test Set
best_rf_model = grid_search.best_estimator_
y_pred_best_rf = best_rf_model.predict(X_test)

print("\nTuned Random Forest Performance:")
print(f"Mean Absolute Error: {mean_absolute_error(y_test, y_pred_best_rf)}")
print(f"Mean Squared Error: {mean_squared_error(y_test, y_pred_best_rf)}")
print(f"R2 Score: {r2_score(y_test, y_pred_best_rf)}")
```
